```python
import pandas as pd
```

```python
players_columns_to_drop = ['Rk', 'Pos', 'FG', 'FGA', 'FG%', '3P',
                    '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%',
                    'FT', 'FTA', 'FT%', 'ORB', 'DRB', 'TRB',
                    'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS']
```

```python
with open("C:\\Users\\andrz\\Documents\\Projects\\NBA\\data\\players_stats\\leagues\\players_stats_2001.csv", 'r', encoding="utf-8") as file:
    df = pd.read_csv(file)

print(df)
```

```python
df_dropped = df.drop(players_columns_to_drop, axis=1)
```

```python
print(df_dropped)
```

```python
df_dropped.rename(columns = {'Player-additional':'ID'}, inplace = True)
```

```python
duplicates_df = df_dropped[df_dropped.duplicated(['ID'], keep=False)]
print(duplicates_df)
```

```python
dropped_duplicates_df = duplicates_df.drop_duplicates(subset='ID', keep="last")
print(dropped_duplicates_df)
```

```python
filtered_df = df_dropped.drop_duplicates(subset='ID', keep="first")
filtered_df = filtered_df.set_index("ID")
filtered_df.head(10)
```

```python
# filtered_df.loc[['alexaco02'], ['Tm']] = dropped_duplicates_df.loc[['alexaco02'], ['Tm']]
filtered_df.loc[['alexaco02'], ['Tm']] = dropped_duplicates_df.query("ID == 'alexaco02'")["Tm"].iloc[0]
filtered_df.head(10)
```
