```python
import pandas as pd
```

```python
all_def = pd.read_csv("./data/league_awards/all_defensive_awards.csv")
```

```python
for index, row in all_def.iterrows():
    if "(T)" in row["Player5"]:
        val = str(row["Player5"]).replace(" (T)", "")
        names = val.split(" ")
        print(names)
        all_def.loc[index, 'Player5'] = " ".join(names[:2])
        all_def.loc[index, 'Player6'] = " ".join(names[2:])
```

```python
all_def['Season'] = all_def['Season'].apply(lambda x: int("20" + str(x)[-2:]))
```

```python
all_def_rows = ["Year", "Player1", "Player2", "Player3", "Player4", "Player5", "Player6", "Player7", "Player8", "Player9", "Player10", "Player11"]
all_def_tranformed = pd.DataFrame(columns=all_def_rows)
all_def_tranformed["Year"] = list(reversed(range(2001, 2024)))
```

```python
all_def_tranformed
```

```python
for index, row in all_def_tranformed.iterrows():
    for i in range(1,12):
        if i <= 5:
            print(all_def_tranformed.loc[index, f"Player{i}"])
            print(row["Year"])
            print(all_def.query(f"Season == '{row['Year']}' and Tm == '1st'")[f"Player{i}"])

            all_def_tranformed.loc[index, f'Player{i}'] = all_def.query(f"Season == '{row['Year']}' and Tm == '1st'")[f"Player{i}"].iloc[0]
        elif 5 < i <= 10:
            all_def_tranformed.loc[index, f'Player{i}'] = all_def.query(f"Season == '{row['Year']}' and Tm == '2nd'")[f"Player{i - 5}"].iloc[0]
        else:
            if pd.isna(all_def.query(f"Season == '{row['Year']}' and Tm == '1st'")[f"Player{i}"].iloc[0]):
                all_def_tranformed.loc[index, f'Player{i}'] = all_def.query(f"Season == '{row['Year']}' and Tm == '1st'")[f"Player{6}"].iloc[0]
            else:
                all_def_tranformed.loc[index, f'Player{i}'] = all_def.query(f"Season == '{row['Year']}' and Tm == '2nd'")[f"Player{6}"].iloc[0]
```
